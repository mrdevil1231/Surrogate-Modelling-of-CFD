```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Assuming X, Y1, and Y2 are defined
Y = np.column_stack((Y1, Y2))  # Multi-output target

# Step 1: Split the data
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)

# Separate targets for Cl and Cd
Cl_train, Cd_train = Y_train[:, 0], Y_train[:, 1]
Cl_test, Cd_test = Y_test[:, 0], Y_test[:, 1]

# Step 2: Scaling input features
scaler = StandardScaler()

# Fit the scaler on the training data and transform both training and test sets
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Step 3: Custom Ridge functions
def ridge_fit(X, Y, alpha):
    # Add bias term
    X_aug = np.hstack([np.ones((X.shape[0], 1)), X])
    n_features = X_aug.shape[1]
    I = np.eye(n_features)
    I[0, 0] = 0  # Don't regularize the bias
    # Closed-form ridge regression
    W = np.linalg.inv(X_aug.T @ X_aug + alpha * I) @ X_aug.T @ Y
    return W

def ridge_predict(X, W):
    X_aug = np.hstack([np.ones((X.shape[0], 1)), X])
    return X_aug @ W

# Step 4: Alpha tuning for Cl
alphas = np.logspace(-3, 2, 20)  # 0.001 to 100
best_alpha_cl = None
lowest_mse_cl = float('inf')
best_W_cl = None
mse_scores_cl = []

for alpha in alphas:
    W_cl = ridge_fit(X_train_scaled, Cl_train, alpha)
    Y_pred_cl = ridge_predict(X_test_scaled, W_cl)
    mse_cl = mean_squared_error(Cl_test, Y_pred_cl)
    mse_scores_cl.append(mse_cl)
    
    if mse_cl < lowest_mse_cl:
        lowest_mse_cl = mse_cl
        best_alpha_cl = alpha
        best_W_cl = W_cl

print(f"\n✅ Best alpha for Cl: {best_alpha_cl:.5f} with MSE: {lowest_mse_cl:.6f}")

# Step 5: Alpha tuning for Cd
best_alpha_cd = None
lowest_mse_cd = float('inf')
best_W_cd = None
mse_scores_cd = []

for alpha in alphas:
    W_cd = ridge_fit(X_train_scaled, Cd_train, alpha)
    Y_pred_cd = ridge_predict(X_test_scaled, W_cd)
    mse_cd = mean_squared_error(Cd_test, Y_pred_cd)
    mse_scores_cd.append(mse_cd)
    
    if mse_cd < lowest_mse_cd:
        lowest_mse_cd = mse_cd
        best_alpha_cd = alpha
        best_W_cd = W_cd

print(f"\n✅ Best alpha for Cd: {best_alpha_cd:.5f} with MSE: {lowest_mse_cd:.6f}")

# Step 6: Predict using best models
Y_pred_cl = ridge_predict(X_test_scaled, best_W_cl)
Y_pred_cd = ridge_predict(X_test_scaled, best_W_cd)

# Step 7: Evaluate Cl and Cd
def evaluate(y_true, y_pred, name):
    print(f"\n{name}")
    print("MSE:", mean_squared_error(y_true, y_pred))
    print("RMSE:", np.sqrt(mean_squared_error(y_true, y_pred)))
    print("MAE:", mean_absolute_error(y_true, y_pred))
    print("R2 score:",r2_score(y_true,y_pred))
    
print("Ridge Regression results using custom functions:\n")
evaluate(Cl_test, Y_pred_cl, "Cl")
evaluate(Cd_test, Y_pred_cd, "Cd")

# Step 8: Optional - Plot Alpha vs MSE for Cl and Cd
plt.figure(figsize=(12, 6))

# Plot for Cl
plt.subplot(1, 2, 1)
plt.plot(alphas, mse_scores_cl, marker='o')
plt.xscale('log')
plt.xlabel('Alpha')
plt.ylabel('MSE')
plt.title('Ridge Regression for Cl: Alpha vs MSE')
plt.grid(True)

# Plot for Cd
plt.subplot(1, 2, 2)
plt.plot(alphas, mse_scores_cd, marker='o')
plt.xscale('log')
plt.xlabel('Alpha')
plt.ylabel('MSE')
plt.title('Ridge Regression for Cd: Alpha vs MSE')
plt.grid(True)

plt.tight_layout()
plt.show()

# Step 9: Combine results
results_df = pd.DataFrame({
    'Cl_true': Cl_test,
    'Cl_predicted': Y_pred_cl,
    'Cd_true': Cd_test,
    'Cd_predicted': Y_pred_cd
})

print("\nPrediction Results:")
print(results_df)
```

### SImple Nueral Network

```python
import numpy as np
from sklearn.model_selection import train_test_split

# ----- Prepare Data -----

# Stack Y1 and Y2 into a single output matrix
Y = np.column_stack((Y1, Y2))

# Separate targets
Y_cl = Y[:, 0]  # Cl
Y_cd = Y[:, 1]  # Cd

# Train-test split for Cl
X_train_cl, X_test_cl, Y_train_cl, Y_test_cl = train_test_split(
    X, Y_cl, test_size=0.2, random_state=None
)

# Train-test split for Cd
X_train_cd, X_test_cd, Y_train_cd, Y_test_cd = train_test_split(
    X, Y_cd, test_size=0.2, random_state=None
)

# Normalize Cl
cl_mean = np.mean(Y_train_cl)
cl_std = np.std(Y_train_cl)
Y_train_cl_scaled = (Y_train_cl - cl_mean) / cl_std
Y_test_cl_scaled = (Y_test_cl - cl_mean) / cl_std

# Normalize Cd
cd_mean = np.mean(Y_train_cd)
cd_std = np.std(Y_train_cd)
Y_train_cd_scaled = (Y_train_cd - cd_mean) / cd_std
Y_test_cd_scaled = (Y_test_cd - cd_mean) / cd_std

# ----- Network Configuration -----

input_size = X.shape[1]
hidden_size = 32
output_size = 1
learning_rate = 0.001
epochs = 100000

# Activation functions
def relu(Z):
    return np.maximum(0, Z)

def relu_derivative(Z):
    return (Z > 0).astype(float)

def mse_loss(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

# ----- Initialize Weights -----

def init_weights(input_size, hidden_size, output_size):
    W1 = np.random.randn(input_size, hidden_size) * 0.01
    b1 = np.zeros((1, hidden_size))
    W2 = np.random.randn(hidden_size, output_size) * 0.01
    b2 = np.zeros((1, output_size))
    return W1, b1, W2, b2

# ----- Training Function -----

def train_network(X_train, Y_train, epochs):
    W1, b1, W2, b2 = init_weights(input_size, hidden_size, output_size)

    for epoch in range(epochs):
        # Forward pass
        Z1 = np.dot(X_train, W1) + b1
        A1 = relu(Z1)
        Z2 = np.dot(A1, W2) + b2
        Y_pred = Z2.squeeze()  # shape (n_samples,)

        # Loss
        loss = mse_loss(Y_train, Y_pred)

        # Backward pass
        dZ2 = (Y_pred - Y_train).reshape(-1,1) / Y_train.shape[0]
        dW2 = np.dot(A1.T, dZ2)
        db2 = np.sum(dZ2, axis=0, keepdims=True)

        dA1 = np.dot(dZ2, W2.T)
        dZ1 = dA1 * relu_derivative(Z1)
        dW1 = np.dot(X_train.T, dZ1)
        db1 = np.sum(dZ1, axis=0, keepdims=True)

        # Update
        W1 -= learning_rate * dW1
        b1 -= learning_rate * db1
        W2 -= learning_rate * dW2
        b2 -= learning_rate * db2

        if epoch % 10000 == 0:
            print(f"Epoch {epoch}, Loss: {loss:.6f}")

    return W1, b1, W2, b2

# ----- Prediction Function -----

def predict(X, W1, b1, W2, b2):
    Z1 = np.dot(X, W1) + b1
    A1 = relu(Z1)
    Z2 = np.dot(A1, W2) + b2
    return Z2.squeeze()

# ----- Train separate networks -----

print("\nTraining model for Cl:")
W1_cl, b1_cl, W2_cl, b2_cl = train_network(X_train_cl, Y_train_cl_scaled, epochs)

print("\nTraining model for Cd:")
W1_cd, b1_cd, W2_cd, b2_cd = train_network(X_train_cd, Y_train_cd_scaled, epochs)

# ----- Predictions -----

# Predict Cl
Y_pred_cl_scaled = predict(X_test_cl, W1_cl, b1_cl, W2_cl, b2_cl)
Y_pred_cl = Y_pred_cl_scaled * cl_std + cl_mean

# Predict Cd
Y_pred_cd_scaled = predict(X_test_cd, W1_cd, b1_cd, W2_cd, b2_cd)
Y_pred_cd = Y_pred_cd_scaled * cd_std + cd_mean

# ----- Final outputs -----

# Actual Cl and Cd for test set
Y_test_cl_true = Y_test_cl
Y_test_cd_true = Y_test_cd

def evaluatecd(y_true, y_pred, name):
    print(f"\n{name}")
    print("MSE:", mean_squared_error(y_true/1000, y_pred/1000))
    print("RMSE:", np.sqrt(mean_squared_error(y_true/1000, y_pred/1000)))
    print("MAE:", mean_absolute_error(y_true/1000, y_pred/1000))
    print("R2 score:",r2_score(y_true,y_pred))

evaluate(Y_test_cl, Y_pred_cl, "Cl")
evaluatecd(Y_test_cd*1000, Y_pred_cd*1000, "Cd")

# Display side-by-side comparison
import pandas as pd

results_df = pd.DataFrame({
    'Cl_true': Y_test_cl_true,
    'Cl_predicted': Y_pred_cl,
    'Cd_true': Y_test_cd_true,
    'Cd_predicted': Y_pred_cd
})

print("\nPrediction Results:")
print(results_df)
```

```python
import numpy as np
from collections import Counter

# Mean Squared Error (MSE) function
def mse(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

# Helper function to calculate best split (renamed to avoid conflict)
def find_best_split(X, y):
    m, n = X.shape
    best_mse = float('inf')
    best_split_info = None
    best_left_X = None
    best_right_X = None
    best_left_y = None
    best_right_y = None

    for feature in range(n):  # Loop over each feature
        thresholds = np.unique(X[:, feature])
        for threshold in thresholds:
            left_mask = X[:, feature] <= threshold
            right_mask = ~left_mask
            left_X, right_X = X[left_mask], X[right_mask]
            left_y, right_y = y[left_mask], y[right_mask]

            if len(left_y) == 0 or len(right_y) == 0:
                continue

            mse_left = mse(left_y, np.mean(left_y))
            mse_right = mse(right_y, np.mean(right_y))
            total_mse = (len(left_y) * mse_left + len(right_y) * mse_right) / len(y)

            if total_mse < best_mse:
                best_mse = total_mse
                best_split_info = (feature, threshold)
                best_left_X, best_right_X = left_X, right_X
                best_left_y, best_right_y = left_y, right_y

    # Return the best split or a default if no valid split is found
    if best_split_info is None:
        return None, None, None, None, None

    return best_split_info, best_left_X, best_right_X, best_left_y, best_right_y

# Helper function to build the decision tree recursively
def build_tree(X, y, max_depth=None, min_samples_split=2, depth=0):
    if len(y) < min_samples_split or (max_depth is not None and depth >= max_depth):
        return np.mean(y)  # Return the mean if we can't split

    # Get the best split using the renamed helper function
    split_result = find_best_split(X, y)
    split_info, left_X, right_X, left_y, right_y = split_result

    if split_info is None:  # If no valid split is found, return the mean
        return np.mean(y)

    feature, threshold = split_info

    left_tree = build_tree(left_X, left_y, max_depth, min_samples_split, depth + 1)
    right_tree = build_tree(right_X, right_y, max_depth, min_samples_split, depth + 1)

    return (feature, threshold, left_tree, right_tree)

# Helper function to predict using the decision tree (unchanged)
def predict_tree(X, tree):
    if isinstance(tree, float):  # Leaf node
        return tree

    feature, threshold, left_tree, right_tree = tree
    if X[feature] <= threshold:
        return predict_tree(X, left_tree)
    else:
        return predict_tree(X, right_tree)


class RandomForest:
    def __init__(self, n_estimators=100, max_depth=None, min_samples_split=2):
        self.n_estimators = n_estimators
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split
        self.trees = []

    def bootstrap_sample(self, X, y):
        n_samples = X.shape[0]
        indices = np.random.choice(range(n_samples), size=n_samples, replace=True)
        return X[indices], y[indices]

    def fit(self, X, y):
        self.trees = []
        for _ in range(self.n_estimators):
            X_bootstrap, y_bootstrap = self.bootstrap_sample(X, y)
            tree = build_tree(X_bootstrap, y_bootstrap, self.max_depth, self.min_samples_split)
            self.trees.append(tree)

    def predict(self, X):
        predictions = np.zeros((X.shape[0], self.n_estimators))
        for i, tree in enumerate(self.trees):
            for j in range(X.shape[0]):
                predictions[j, i] = predict_tree(X[j], tree)
        return np.mean(predictions, axis=1)

# ----- The rest of your code remains unchanged -----

# ----- Prepare Data -----
# Example data (X and Y1, Y2 should be already defined in your context)
Y = np.column_stack((Y1, Y2))  # Stack Y1 (Cl) and Y2 (Cd)

# Convert to NumPy arrays (if they are DataFrames)
X = np.array(X)  # Convert X to NumPy array if it is a DataFrame
Y = np.array(Y)  # Convert Y to NumPy array if it is a DataFrame

# Separate targets
Y_cl = Y[:, 0]  # Cl
Y_cd = Y[:, 1]  # Cd

# Train-test split for Cl
X_train_cl, X_test_cl, Y_train_cl, Y_test_cl = train_test_split(X, Y_cl, test_size=0.3, random_state=None)

# Train-test split for Cd
X_train_cd, X_test_cd, Y_train_cd, Y_test_cd = train_test_split(X, Y_cd, test_size=0.3, random_state=None)

# ----- Normalize Data -----
cl_mean = np.mean(Y_train_cl)
cl_std = np.std(Y_train_cl)
Y_train_cl_scaled = (Y_train_cl - cl_mean) / cl_std
Y_test_cl_scaled = (Y_test_cl - cl_mean) / cl_std

cd_mean = np.mean(Y_train_cd)
cd_std = np.std(Y_train_cd)
Y_train_cd_scaled = (Y_train_cd - cd_mean) / cd_std
Y_test_cd_scaled = (Y_test_cd - cd_mean) / cd_std

# ----- Train Random Forest for Cl -----
rf_cl = RandomForest(n_estimators=100, max_depth=5)
rf_cl.fit(X_train_cl, Y_train_cl_scaled)

# ----- Train Random Forest for Cd -----
rf_cd = RandomForest(n_estimators=100, max_depth=5)
rf_cd.fit(X_train_cd, Y_train_cd_scaled)

# ----- Predict Cl and Cd -----
Y_pred_cl_scaled = rf_cl.predict(X_test_cl)
Y_pred_cl = Y_pred_cl_scaled * cl_std + cl_mean  # Denormalize

Y_pred_cd_scaled = rf_cd.predict(X_test_cd)
Y_pred_cd = Y_pred_cd_scaled * cd_std + cd_mean  # Denormalize



print("Using Custom Fuctions: ")
evaluate(Y_test_cd, Y_pred_cd,'Cl')
evaluatecd(Y_test_cd, Y_pred_cd,'Cd')

results_df = pd.DataFrame({
    'Cl_true': Y_test_cl,
    'Cl_predicted': Y_pred_cl,
    'Cd_true': Y_test_cd,
    'Cd_predicted': Y_pred_cd
})

print("\nPrediction Results:")
print(results_df)
```
